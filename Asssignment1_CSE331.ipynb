```python
##4
max_age = df['age'].max()
min_age = df['age'].min()
mean_age = df['age'].mean()

print(f"Maximum age: {max_age}")
print(f"Minimum age: {min_age}")
print(f"Mean age: {mean_age}")
```

```python
##5
df['full_name'] = df['first'] + ' ' + df['last']
common_names = df['full_name'].value_counts().head(3)
print(f"Three most common customer names are:\n {common_names}")
```

```python
##6
same_numbers = df[df.duplicated('phone', keep = False)]
print(same_numbers[['first', 'last', 'phone']])
```

```python
##7
count = len(df[df['profession'] == 'Structural Engineer'])
print(f"Number of structural engineers= {count}")
```

```python
##8
male_str_eng = df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Male')]
print(f"Number of male structural engineers= {len(male_str_eng)}")
```

```python
##9
female_str_eng_ab = df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Female') & (df['province'] == 'AB')]
print(f"Female structural engineers from the province Alberta= {len(female_str_eng_ab)}")
```

```python
##10
max_spending = df['price(CAD)'].max()
min_spending = df['price(CAD)'].min()
avg_spending = df['price(CAD)'].mean()

print(f"Maximum spending: {max_spending}")
print(f"Minimum spending: {min_spending}")
print(f"Average spending: {avg_spending}")
```

```python
##11
no_spending = df[df['price(CAD)'] == 0][['first', 'last', 'price(CAD)']]
print(f"Customers who did not spend anything:\n {no_spending}")
```

```python
##12
hundred_plus_spending = df[df['price(CAD)'] >= 100][['first', 'last', 'price(CAD)']]
print(f"person who Spent 100CAD or more:\n {hundred_plus_spending}")
```

```python
##13
emails = df[df['cc_no'] == 5020000000000230][['email']]
print(f"Number of emails that are associated with this credit card number ‘5020000000000230’: {len(emails)}")
```

```python
##14
expires_19 = df[df['cc_exp'].str.endswith('2019')]
print(f"number of cards that are expiring in 2019: {len(expires_19)}")
```

```python
##15
visa_card = df[df['cc_type'] == "Visa"]
print(f"Number of people use visa card: {len(visa_card)}")
```

```python
##16
name = df[(df['cc_type'] == "Visa") & (df['price(CAD)'] == 100)][['first', 'last', 'price(CAD)']]
print(f"The customer spent 100 CAD using visa is:\n {name}")
```

```python
##17
common_proff = df['profession'].value_counts().head(2)
print(f"two most common professions are: \n {common_proff}")
```

```python
##18
df['provider'] = df['email'].str.split("@").str[1]
top_5 = df['provider'].value_counts().head(5)
print(f"Tof 5 email provider: \n {top_5}")
```

```python
##19
am_edu_mail = df[df['email'].str.contains('am.edu')][['first', 'last', 'email']]
print(f"Customer with am.edu mail: \n {am_edu_mail}")
```

```python
##20
most_common_day = df['weekday'].value_counts().head(1)
print(f"The store gets most customers on:\n {most_common_day}")
```
